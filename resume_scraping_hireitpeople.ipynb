```python
import requests
from bs4 import BeautifulSoup
import pickle
```

```python
Links = list()
for i in range(1,199):
    url = 'https://www.hireitpeople.com/resume-database/76-network-admin-resumes/page/'+str(i)
    response = requests.get(url)
    soup = BeautifulSoup(response.content)
    title_tags = soup.find('table',attrs = {'class':'table table-border hit-table'})
    title_tags1 = title_tags.findAll('a')

    for title_tag in title_tags1:
            Links.append(title_tag['href'])

links = []
for link in Links:
    if link not in links:
        links.append(link)
        
```

```python
results = []
for link1 in links:
    res = requests.get(link1)
    print(res.status_code, link1)
    content = BeautifulSoup(res.content)
    pickle.dump({content.find('div', attrs={'class': 'single-post-body'}).get_text()},open(link1[70:76] +'.txt','wb'))
    
    
    
```
